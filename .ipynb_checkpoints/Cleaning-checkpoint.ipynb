```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url='https://transcripts.foreverdreaming.org/viewtopic.php?f=177&t'
def func(f):
    df=pd.DataFrame()
    for i in range(505,713):
        try:
            url=str(f)+f'{i}'
        except:
            None

        html = requests.get(url)
        soup= BeautifulSoup(html.content, "html.parser")
        h=soup.findAll('p')[2:-5]
        x= soup.findAll("h2")[0].getText()
        H=[i.getText().split(':') for i in h]
        Lista=[]
        try:
            for i in H:
                if len(i) ==2:
                    Lista.append(i)

            for i in Lista:
                if '(' in i[0]:
                    i[0].split(')')[1]
        except:
            'IndexError: list index out of range'
        df1=pd.DataFrame(Lista)
        df1['temporada']= x[0:2]
        df1['episodio']=x[3:5]
        df1['nombre_episodio']=x.split('-')[1]
        df= pd.concat([df,df1])
    return df
```

```python
df3=func('https://transcripts.foreverdreaming.org/viewtopic.php?f=177&t=11')
```

```python
df3
```

```python
df3['frase']= df3[1]
df3['personaje']=df3[0]
df3.drop(1, inplace=True, axis=1)
df3.drop(0, inplace=True, axis= 1)
```

```python
def change(x):
    if x == 'Future Ted VO': #
        return 'Future Ted(Narrator)'
    elif x== 'Ted (2030)':
        return 'Future Ted(Narrator)'
    elif x== 'Ted (voix off)':
        return 'Future Ted(Narrator)'
    elif x== 'Ted from 2030':
        return 'Future Ted(Narrator)'
    elif x== 'NARRATOR':
        return 'Future Ted(Narrator)'
    elif x== 'Narrator':
        return 'Future Ted(Narrator)'
    elif x== 'BARNEY':
        return 'Barney'
    elif x== 'TED':
        return 'Ted'
    elif x== 'MARSHALL':
        return 'Marshall'
    elif x== 'ROBIN':
        return 'Robin'
    elif x== 'LILY':
        return 'Lily'
    elif x== 'Waitress':
        return 'Wendy'
    else:
        return x
```

```python
df4=df3
```

```python
df4['personaje']= df4['personaje'].apply(change)
```

```python
df4 = df4[df4.groupby('personaje').personaje.transform(len) > 20]
```
